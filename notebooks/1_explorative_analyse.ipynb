```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# üìÇ Daten laden (Pfad anpassen falls n√∂tig)
df_train = pd.read_csv("../data/set-A_train/image_labels.csv")

### üîç 1. Leere vs. nicht-leere Bilder
empty_counts = df_train["Is_nonempty"].value_counts()
print("\nüìä Anzahl leerer vs. nicht-leerer Bilder:")
print(empty_counts)

plt.figure(figsize=(6,4))
sns.countplot(x=df_train["Is_nonempty"])
plt.title("Leere vs. nicht-leere Bilder")
plt.xlabel("Ist ein Schiff im Bild?")
plt.ylabel("Anzahl Bilder")
plt.show()


### üîç 2. Verteilung der Schiffstypen
ship_counts = df_train["Ship"].value_counts()
print("\nüö¢ Verteilung der Schiffstypen:")
print(ship_counts)

plt.figure(figsize=(8,4))
sns.countplot(y=df_train["Ship"], order=ship_counts.index)
plt.title("Verteilung der Schiffstypen")
plt.xlabel("Anzahl Bilder")
plt.ylabel("Schiffstyp")
plt.show()


### üîç 3. Verteilung der Schiffspositionen
location_counts = df_train["Location"].value_counts()
print("\nüìç Verteilung der Schiffspositionen:")
print(location_counts)

plt.figure(figsize=(8,6))
sns.countplot(y=df_train["Location"], order=location_counts.index)
plt.title("Verteilung der Schiffslokationen")
plt.xlabel("Anzahl Bilder")
plt.ylabel("Position auf dem Spielfeld")
plt.show()


### üîç 4. Richtung der Schiffe (West vs. East)
heading_counts = df_train["Heading"].value_counts()
print("\nüß≠ Verteilung der Schiffsausrichtung (West vs. East):")
print(heading_counts)

plt.figure(figsize=(6,4))
sns.countplot(x=df_train["Heading"], order=["West", "East"])
plt.title("Schiffsausrichtung")
plt.xlabel("Richtung des Schiffs")
plt.ylabel("Anzahl Bilder")
plt.show()
```
